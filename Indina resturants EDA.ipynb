# EDA
 ### Restaurant Chains
     Chains vs Outlets
     Top Restaurant Chains (by number of outlets)
     Top Restaurant Chains (by average ratings)



```python
distinct_shopes=df['name'].value_counts()
print(distinct_shopes)
chains=distinct_shopes[distinct_shopes>=2]
```

```python
total_shopes=df['name'].shape[0]
print("total noumber of shopes=",total_shopes)
print("Total number of restraunt chains=",chains.shape[0])
print("total nuber of restutants which are part of some chain=",total_shopes-distinct_shopes[distinct_shopes==1].shape[0])
print("percentage of resturant chain=",(total_shopes-distinct_shopes[distinct_shopes==1].shape[0])/total_shopes *100,"%")
```

```python
### top resturants by chain are 
print(chains)
```

```python
y=chains[:10].index
x=chains[:10]
plt.barh(y,x)
plt.title('top 10 resturants by number of outlets')
plt.xlabel('outlets')
plt.ylabel('')
plt.show()
```

```python
## Top 10 resturants based on average rating 
rating_based=df[df['name'].isin(chains.index)].groupby('name').mean().sort_values(by='aggregate_rating',ascending=False)['aggregate_rating'][:10]
rating_based
```

```python
x=rating_based
y=rating_based.index
plt.figure(figsize=(10, 8))
plt.barh(y,x)
plt.title('top 20 resturants by average rating')
plt.xlabel('rating')
plt.ylabel('')
for i, v in enumerate(x):
    plt.text(v, i, str(round(v, 2)), color='black')

plt.show()
```
